````scala
def length[A](xs: List[A]): Int = {
    def len[A](yys: List[A], acc: Int): Int = yys match {
        case Nil   => acc
        case y::ys => len(ys, acc + 1)
    }
    len(xs, 0)
}

length(List.range(0, 10))
````

## Optimize reversing lists

Given the following reversing list implementation,

```scala
def reverse[A](xs: List[A]): List[A] = xs match {
    case Nil     => Nil
    case x :: xs => reverse(xs) ::: List(x)
}
```

which time complexity is $O(n(n + 1)/2)$, implement a more efficient algorithm.

PSS. the idea of accumulate and reduce may help you! :)

````scala
// code goes here...
````

````scala
def reverse[A](xs: List[A]): List[A] = {
    (xs foldLeft List[A]()) {(acc: List[A], elem: A) => elem :: acc }
}

reverse(List.range(0, 10))
````

````scala
def mapFun[A, B](xs: List[A], f: A => B): List[B] = 
    (xs foldRight List[B]()){ (x: A, acc: List[B]) => f(x) :: acc }

mapFun(List.range(0, 10), (x => x * x): Int => Int)
````

````scala
def filterFun[A](xs: List[A], p: A => Boolean): List[A] = {
    (xs foldLeft List[A]()) { (acc: List[A], x: A) => if (p(x)) acc ++ List(x) else acc } // if you see this, better use foldRight :)
}
def isEven(x: Int): Boolean = x % 2 == 0

filterFun(List.range(0, 10), isEven)
````
